```python
print("\n-- Exercise 1.1:Temperature converter")
#function to convert celsius to fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32
#function to convert fahrenheit to celsius
def fahrenheit_to_celsius(fahrenheit):
    return (fahrenheit - 32) * 5/9
#test values
celsius_value = 25
fahrenheit_value = 77
#Testing the functions
print(f"{celsius_value}°c =  {celsius_to_fahrenheit(celsius_value)}°f")
print(f"{fahrenheit_value}°f = {fahrenheit_to_celsius(fahrenheit_value)}°c")
```

```python
print("\n--- Exercise 1.2: String Analyzer ---")

# Function to analyze a given string
def analyze_string(text):
    vowels_set = "aeiouAEIOU"

    # Total number of characters in the string (including spaces)
    length = len(text)
    vowels = 0
    consonants = 0

    # Loop through each character in the text
    for ch in text:

        # Check if the character is an alphabet letter
        if ch.isalpha():
            if ch in vowels_set:
                vowels += 1
            else:
                consonants += 1

    words = len(text.split())

    # Return results as a dictionary
    return {
        "length": length,
        "vowels": vowels,
        "consonants": consonants,
        "words": words
    }
  #Test
result = analyze_string("Python programming is amazing")
print(result)
```

```python
# Exercise 1.3: Calculator Function
print("\n--- Exercise 1.3: Calculator Function ---")

def calculator(num1, num2, operation="+"):
    try:
        if operation == "+":
            return num1 + num2
        elif operation == "-":
            return num1 - num2
        elif operation == "*":
            return num1 * num2
        elif operation == "/":
            if num2 == 0:
                return "Error: Division by zero"
            return num1 / num2
        elif operation == "**":
            return num1 ** num2
        elif operation == "%":
            if num2 == 0:
                return "Error: Modulo by zero"
            return num1 % num2
        else:
            return "Error: Unsupported operation"
    except Exception as e:
        return f"Error: {e}"


# Test with various operations
print("Addition:", calculator(10, 5))
print("Subtraction:", calculator(10, 5, "-"))
print("Multiplication:", calculator(10, 5, "*"))
print("Division:", calculator(11, 5, "/"))
print("Power:", calculator(2, 3, "**"))
print("Modulo:", calculator(10, 3, "%"))
print("Division by zero:", calculator(10, 0, "/"))
```

```python
# Exercise 2.2: List Statistics
print("\n--- Exercise 2.2: List Statistics ---")

# Function to calculate statistics of a list of numbers
def list_statistics(numbers):
    # Create a dictionary to store results
    stats = {}

    # Calculate sum of all numbers
    stats["sum"] = sum(numbers)

    # Calculate average of numbers
    stats["average"] = stats["sum"] / len(numbers)

    # Find the smallest number
    stats["minimum"] = min(numbers)

    # Find the largest number
    stats["maximum"] = max(numbers)

    # Calculate the range (max - min)
    stats["range"] = stats["maximum"] - stats["minimum"]

    # Count number of elements in the list
    stats["count"] = len(numbers)

    # Return the dictionary containing all statistics
    return stats


# Test the function with given list
numbers = [15, 23, 8, 42, 16, 4, 38, 27]
result = list_statistics(numbers)

# Print the result
print("List statistics:")
print(result)
```

```python
# Exercise 2.4: Nested Lists
print("\n--- Exercise 2.4: Nested Lists ---")

# Given matrix (nested list)
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

#Print the entire matrix (one row per line)
print("Matrix:")
for row in matrix:
    print(row)

# Access and print the element at row 1, column 2
# Indexing starts from 0 → row 1 means second row, column 2 means third column
print("Element at row 1, column 2:", matrix[1][2])
# Calculate and print the sum of all elements
total_sum = 0
for row in matrix:
    for value in row:
        total_sum += value

print("Sum of all elements:", total_sum)

# Create a flattened version (1D list) of the matrix
flattened = [value for row in matrix for value in row]
print("Flattened list:", flattened)

# Create the transpose of the matrix
# Rows become columns
transpose = []
for i in range(len(matrix[0])):
    new_row = []
    for row in matrix:
        new_row.append(row[i])
    transpose.append(new_row)

print("Transpose of the matrix:")
for row in transpose:
    print(row)
```

```python
# Exercise 2.5: List Sorting and Searching
print("\n--- Exercise 2.5: List Sorting and Searching ---")

# Function to find duplicate elements
def find_duplicates(lst):
    duplicates = []
    for item in lst:
        if lst.count(item) > 1 and item not in duplicates:
            duplicates.append(item)
    return duplicates


# Function to remove duplicates (keep order)
def remove_duplicates(lst):
    unique = []
    for item in lst:
        if item not in unique:
            unique.append(item)
    return unique


# Test list
numbers = [1, 2, 3, 2, 4, 5, 3, 6, 7, 5]

# Find duplicates
print("Duplicate elements:", find_duplicates(numbers))

# Remove duplicates
print("List without duplicates:", remove_duplicates(numbers))

# Sort the list
print("Sorted list:", sorted(numbers))
```

```python
# Exercise 3.2: Word Frequency Counter
print("\n--- Exercise 3.2: Word Frequency Counter ---")

# Function to count word frequency
def word_frequency(text):
    frequency = {}  # dictionary to store word counts

    # Convert text to lowercase
    text = text.lower()

    # Remove punctuation by replacing it with empty space
    punctuation = ".,!?;:"
    for p in punctuation:
        text = text.replace(p, "")

    # Split the text into words
    words = text.split()

    # Count each word
    for word in words:
        if word in frequency:
            frequency[word] += 1
        else:
            frequency[word] = 1

    return frequency


# Test the function
text = "Python is great. Python is powerful. Python is easy to learn."
result = word_frequency(text)

# Print the result
print("Word frequency:")
print(result)
```

```python
# Exercise 3.4: Nested Dictionaries
print("\n--- Exercise 3.4: Nested Dictionaries ---")

# Inventory dictionary (nested dictionaries)
inventory = {
    "fruits": {
        "apple": {"price": 0.5, "stock": 100},
        "banana": {"price": 0.3, "stock": 150},
        "orange": {"price": 0.6, "stock": 80}
    },
    "vegetables": {
        "carrot": {"price": 0.4, "stock": 120},
        "broccoli": {"price": 1.2, "stock": 50}
    }
}

# Print all items with their prices and stock
print("Inventory details:")

for category in inventory:                     # fruits, vegetables
    print("\nCategory:", category)
    for item in inventory[category]:           # apple, banana, etc.
        price = inventory[category][item]["price"]
        stock = inventory[category][item]["stock"]
        print(item, "- Price:", price, "Stock:", stock)

# Calculate total inventory value
total_value = 0

for category in inventory:
    for item in inventory[category]:
        price = inventory[category][item]["price"]
        stock = inventory[category][item]["stock"]
        total_value += price * stock

print("\nTotal inventory value:", total_value)
```

```python
print("\n--- Exercise 3.5: Dictionary Operations ---")

# Function to merge dictionaries
def merge_dictionaries(dict1, dict2, dict3):
    merged = {}          # empty dictionary

    merged.update(dict1)
    merged.update(dict2)
    merged.update(dict3)

    return merged


# Function to invert a dictionary
def invert_dictionary(d):
    inverted = {}        # empty dictionary

    for key in d:
        value = d[key]
        inverted[value] = key

    return inverted


# Test dictionaries
dict1 = {"a": 1, "b": 2}
dict2 = {"c": 3, "d": 4}
dict3 = {"e": 5, "f": 6}

# Merge dictionaries
merged_dict = merge_dictionaries(dict1, dict2, dict3)
print("Merged Dictionary:", merged_dict)

# Invert dictionary
inverted_dict = invert_dictionary(merged_dict)
print("Inverted Dictionary:", inverted_dict)
```

```python
print("\n--- Exercise 4.1: Student Management System ---")
#Add a student
def add_student(students_list, name, age, grades):
    average = sum(grades) / len(grades)   # calculate average

    student = {                           # student dictionary
        "name": name,
        "age": age,
        "grades": grades,
        "average": average
    }

    students_list.append(student)
#Get top n students by average
def get_top_students(students_list, n=3):
    sorted_students = sorted(
        students_list,
        key=lambda s: s["average"],
        reverse=True
    )
    return sorted_students[:n]
#Get statistics
def get_statistics(students_list):
    averages = []

    for student in students_list:
        averages.append(student["average"])

    stats = {
        "total_students": len(students_list),
        "overall_average": sum(averages) / len(averages),
        "highest_average": max(averages),
        "lowest_average": min(averages)
    }
    return stats
#  Filter students by minimum average
def filter_by_average(students_list, min_average):
    result = []

    for student in students_list:
        if student["average"] >= min_average:
            result.append(student)

    return result
# testing
students = []

add_student(students, "Veda", 20, [85, 90, 88])
add_student(students, "Parimala", 21, [70, 75, 72])
add_student(students, "Jyothi", 19, [95, 92, 96])
add_student(students, "Varshni", 22, [60, 65, 63])
add_student(students, "Sravya", 20, [88, 84, 90])

print("\nAll Students:")
for s in students:
    print(s)

print("\nTop Students:")
for s in get_top_students(students):
    print(s["name"], "-", s["average"])

print("\nStatistics:")
print(get_statistics(students))

print("\nStudents with average >= 85:")
for s in filter_by_average(students, 85):
    print(s["name"], "-", s["average"])
```

```python
print("\n--- Exercise 4.2: Data Transformation Pipeline ---")

# Function to transform data
def transform_data(data):
    transformed = []   #  list to store updated records

    for person in data:
        # Create full name
        full_name = person["first_name"] + " " + person["last_name"]

        # Decide age group
        age = person["age"]
        if age < 18:
            age_group = "Minor"
        elif age <= 24:
            age_group = "Young Adult"
        elif age <= 39:
            age_group = "Adult"
        else:
            age_group = "Senior"

        # Decide performance level
        score = person["score"]
        if score >= 90:
            performance = "Excellent"
        elif score >= 75:
            performance = "Good"
        elif score >= 60:
            performance = "Average"
        else:
            performance = "Needs Improvement"

        # Add new data to dictionary
        person["full_name"] = full_name
        person["age_group"] = age_group
        person["performance_level"] = performance

        # Add updated person to new list
        transformed.append(person)

    return transformed


# Test data
raw_data = [
    {"first_name": "John", "last_name": "Doe", "age": 22, "score": 85},
    {"first_name": "Jane", "last_name": "Smith", "age": 28, "score": 92},
    {"first_name": "Bob", "last_name": "Johnson", "age": 19, "score": 78},
    {"first_name": "Alice", "last_name": "Williams", "age": 35, "score": 95}
]

# Run transformation
result = transform_data(raw_data)

# Print result
for item in result:
    print(item)
```
