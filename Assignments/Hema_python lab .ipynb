```python
# exercise 3: Arithmetic Operations and Calculations
# Take user input
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
#calculations
addition = num1 + num2
subtraction = num1 - num2
multiplication = num1 * num2
division = num1 / num2
floor_division = num1 // num2
modulus = num1 % num2
exponentiation = num1 ** num2
# Display results
print("\n" + "=" * 60)
print("Calculator Results")
print("=" * 60)
print(f"Addition: {addition}")
print(f"Subtraction: {subtraction}")
print(f"Multiplcation:{multiplication}")
print(f"Division:{division}")
print(f"Floor Division:{floor_division}")
print(f"Modulus:{modulus}")
print(f"Exponentiation:{exponentiation}")
```

```python
# exercise 4: String Manipulation
# user input
sentence = input("Enter a sentence: ")
old_word = input("Enter a word to replace: ")
new_word = input("Enter the new word: ")
# String operations
uppercase = sentence.upper()
lowercase = sentence.lower()
title_case = sentence.title()
# Counts
character_count = len(sentence)
word_count = len(sentence.split())
modified_sentence = sentence.replace(old_word, new_word) # Replace word
# Display results
print("\n" + "=" * 60)
print("String Analysis")
print("=" * 60)
print(f"Original: {sentence}")
print(f"Uppercase: {uppercase}")
print(f"Lowercase: {lowercase}")
print(f"Title Case: {title_case}")
print(f"Character count: {character_count}")
print(f"Word count: {word_count}")
print(f"Modified: {modified_sentence}")
```

```python
# exercise 5: Temperature Converter
celsius = float(input("Enter temperature in Celsius: "))
# Conversions
fahrenheit = (celsius * 9/5) + 32
kelvin = celsius + 273.15
# Display results
print("\n" + "=" * 60)
print("Temperature Conversion")
print("=" * 60)
print(f"Celsius:    {celsius:.2f}°C")
print(f"Fahrenheit: {fahrenheit:.2f}°F")
print(f"Kelvin:     {kelvin:.2f}K")
```

```python
# exercise 6: Variable Swapping and Multiple Assignment
# Multiple assignment with the same value
a = b = c = 10
print("Multiple assignment with same value: ")
print(f"a = {a}, b = {b}, c = {c}\n")
# different values
x, y, z = 5, 10, 15
print("Multiple assignment (different values):")
print(f"x = {x}, y = {y}, z = {z}\n")
#Swapping two variables without a temporary variable
print("Before swap:")
print(f"x = {x}, y = {y}")
x, y = y, x  # swap using multiple assignment
print("After swap:")
print(f"x = {x}, y = {y}")
```

```python
# exercise 7 :shopping cart calculator
print("Enter product information:")
# Product 1,2,3
product1_name = input("Product 1 name: ")
product1_price = float(input("Product 1 price: $"))
product2_name = input("Product 2 name: ")
product2_price = float(input("Product 2 price: $"))
product3_name = input("Product 3 name: ")
product3_price = float(input("Product 3 price: $"))
# Calculate subtotal
subtotal = product1_price + product2_price + product3_price
if subtotal > 100:
    discount = subtotal * 0.10   # 10% discount
else:
    discount = 0
# After discount
price_after_discount = subtotal - discount
# Calculate tax (8%)
tax = price_after_discount * 0.08
# Final total
total = price_after_discount + tax
# Display receipt
print("\nSHOPPING CART RECEIPT")
print("-------")
print(product1_name, "-", product1_price)
print(product2_name, "-", product2_price)
print(product3_name, "-", product3_price)
print("-------")
print("Subtotal:", subtotal)
if discount > 0:
    print("Discount:", discount)
    print("After Discount:", price_after_discount)
print("Tax:", tax)
print("Total:", total)
```

```python
#Input Validation -- challenge 1
while True:
    try:
        # Take input from the user
        num = float(input("Enter a positive number: "))

        if num > 0:
            # If valid, display the number
            print("Valid input:", num)
            # Exit the loop
            break
        else:
            print("Number must be positive.")

    # block runs if the input cannot be converted to a number
    except ValueError:
        print("Invalid input! Enter numbers only.")
```

```python
#challenge 2 : Enhanced Calculator with menu options
while True:
    print("\n--- Calculator Menu ---")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    print("5. Exit")
    # user's choice
    choice = input("Choose an operation (1-5): ")
    # Exit condition
    if choice == "5":
        print("Calculator closed.")
        break
    try:
        # Taking  two numbers from user
        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))
        if choice == "1":
            print("Result:", num1 + num2)
        elif choice == "2":
            print("Result:", num1 - num2)
        elif choice == "3":
            print("Result:", num1 * num2)
        elif choice == "4":
            if num2 != 0:
                print("Result:", num1 / num2)
            else:
                print("Cannot divide by zero")
        else:
            print("Invalid choice. Please select 1-5.")
    except ValueError:
        print("Invalid input! Enter numbers only.")
```

```python
# challeng :3 Calculate change after payment
# Total bill amount
total_amount = float(input("Enter total bill amount: ₹"))
while True:
    try:
        # Cash given by customer
        cash_paid = float(input("Enter cash paid by customer: ₹"))
        # Check if cash is enough
        if cash_paid >= total_amount:
            change = cash_paid - total_amount
            print("\n--- Receipt ---")
            print(f"Total Bill: ₹{total_amount:.2f}")
            print(f"Cash Paid: ₹{cash_paid:.2f}")
            print(f"Change : ₹{change:.2f}")
            break
        else:
            print(" Cash is less than total bill. Please enter sufficient amount.")

    except ValueError:
        print("Invalid input! Enter numbers only.")
```

```python
# Challenge 5 :temperature converter
# Temperature Unit Converter

while True:
    print("\n--- Temperature Converter ---")
    print("1. Celsius to Fahrenheit")
    print("2. Fahrenheit to Celsius")
    print("3. Celsius to Kelvin")
    print("4. Kelvin to Celsius")
    print("5. Exit")

    choice = input("Choose an option (1-5): ")

    if choice == "5":
        print("Converter closed.")
        break

    try:
        temp = float(input("Enter temperature value: "))

        if choice == "1":
            result = (temp * 9/5) + 32
            print("Temperature in Fahrenheit:", result)

        elif choice == "2":
            result = (temp - 32) * 5/9
            print("Temperature in Celsius:", result)

        elif choice == "3":
            result = temp + 273.15
            print("Temperature in Kelvin:", result)

        elif choice == "4":
            result = temp - 273.15
            print("Temperature in Celsius:", result)

        else:
            print("Invalid choice. Select 1-5.")

    except ValueError:
        print("Invalid input! Enter numbers only.")
```
