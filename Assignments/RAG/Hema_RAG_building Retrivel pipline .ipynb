````python
def chunk_by_words(text, max_words=50, overlap=10):
# -------------------------------
# Step 1: Validate input values
# -------------------------------
# max_words must be positive
    if max_words <= 0:
        raise ValueError("max_words must be greater than 0")
# overlap must be smaller than max_words
# otherwise the sliding window will not move properly
    if overlap >= max_words:
        raise ValueError("overlap must be smaller than max_words")
# -------------------------------
# Step 2: Split text into words
# -------------------------------
# Split text into a list of words using whitespace
    words = text.split()
# List to store final chunks
    chunks = []
# -------------------------------
# Step 3: Define sliding window
# -------------------------------
# Start index of first chunk
    start = 0
# We subtract overlap so that chunks share some words
    step = max_words - overlap
# -------------------------------
# Step 4: Create chunks
# -------------------------------
# Continue until we reach the end of word list
    while start < len(words):
# Define end of current chunk
        end = start + max_words
# Slice words from start to end
        chunk = words[start:end]
# Join words back into a string
        chunks.append(" ".join(chunk))
# Move window forward
        start += step
# -------------------------------
# Step 5: Return all chunks
# -------------------------------
    return chunks
# ==========================================
# TEST THE FUNCTION
# ==========================================
sample_text = (
    "Machine learning is a subset of artificial intelligence. "
    "It enables systems to learn from experience. "
    "Deep learning uses neural networks with multiple layers."
)
# Create chunks with 10 words per chunk
# and 3 words overlapping between chunks
chunks = chunk_by_words(sample_text, max_words=10, overlap=3)
# Print number of chunks
print(f"Number of chunks: {len(chunks)}\n")
# Print each chunk
for i, chunk in enumerate(chunks):
    print(f"Chunk {i+1}: {chunk}")
````

````python
def chunk_by_sentences(text, max_sentences=2):
# ------------------------------------
# Step 1: Validate input
# ------------------------------------
    if max_sentences <= 0:
        raise ValueError("max_sentences must be greater than 0")
# ------------------------------------
# Step 2: Split text into sentences
# ------------------------------------
# Simple sentence splitting based on ". "
# This assumes sentences end with a period followed by a space
    sentences = text.split(". ")
# remove extra whitespace and ignore empty strings
    sentences = [s.strip() for s in sentences if s.strip()]
# ------------------------------------
# Step 3: Group sentences into chunks
# ------------------------------------
    chunks = []
# Loop through sentences in steps of max_sentences
    for i in range(0, len(sentences), max_sentences):
# Select group of sentences
        chunk_sentences = sentences[i:i + max_sentences]
# Join them back into a paragraph
        chunk = ". ".join(chunk_sentences)
# Ensure the chunk ends with a period
        if not chunk.endswith("."):
            chunk += "."

        chunks.append(chunk)
# ------------------------------------
# Step 4: Return final chunks
# ------------------------------------
    return chunks
# ==========================================
# TEST THE FUNCTION
# ==========================================

sample_text = (
    "Machine learning is a subset of artificial intelligence. "
    "It enables systems to learn from experience. "
    "Deep learning uses neural networks with multiple layers. "
    "Neural networks are inspired by biological systems."
)

chunks = chunk_by_sentences(sample_text, max_sentences=2)

print(f"Number of chunks: {len(chunks)}\n")

for i, chunk in enumerate(chunks):
    print(f"Chunk {i+1}: {chunk}")
````

````python
# Import Libraries
import numpy as np
from sentence_transformers import SentenceTransformer
import warnings
import os
# Suppress warnings and progress bars
warnings.filterwarnings("ignore")
os.environ["TRANSFORMERS_VERBOSITY"] = "error"
os.environ["HF_HUB_DISABLE_PROGRESS_BARS"] = "1"
# Sample Knowledge Base
DOCUMENTS = [
    "Machine learning is a method where computers learn from data.",
    "Deep learning is a subset of machine learning that uses neural networks.",
    "Natural language processing helps computers understand human language.",
    "Computer vision enables machines to interpret images and videos.",
    "Reinforcement learning is a type of machine learning based on rewards."
]
# Initialize Embedding Model
print("Loading model...")
model = SentenceTransformer('all-MiniLM-L6-v2')
# Create Document Embeddings
print("Generating document embeddings...")
embeddings = model.encode(DOCUMENTS)
# Cosine Similarity Function
def cosine_similarity(vec1, vec2):
    return np.dot(vec1, vec2) / (
        np.linalg.norm(vec1) * np.linalg.norm(vec2)
    )
# Retrieval Function
def retrieve_documents(query, documents, embeddings, model, top_k=3):
    # Encode query
    query_embedding = model.encode(query)

    # Compute similarity with each document
    similarities = []

    for i, doc_embedding in enumerate(embeddings):
        score = cosine_similarity(query_embedding, doc_embedding)
        similarities.append((documents[i], score))

    #Sort by similarity (highest first)
    similarities.sort(key=lambda x: x[1], reverse=True)

    #Return top_k results
    return similarities[:top_k]
# Test Retrieval
query = "How do computers learn from data?"
results = retrieve_documents(query, DOCUMENTS, embeddings, model, top_k=3)
print("\nQuery:", query, "\n")
for i, (doc, score) in enumerate(results, 1):
    print(f"Result {i} (Score: {score:.4f}):")
    print(doc)
    print()
````

### Task 4.1: Implement Batch Processing

Create a function that efficiently handles multiple queries at once.


````python
# COMPLETE BATCH RETRIEVAL SYSTEM
import numpy as np
from sentence_transformers import SentenceTransformer
import warnings
import os
# Suppress warnings
warnings.filterwarnings("ignore")
os.environ["TRANSFORMERS_VERBOSITY"] = "error"
os.environ["HF_HUB_DISABLE_PROGRESS_BARS"] = "1"
# Knowledge Base
# --------------
DOCUMENTS = [
    "Machine learning is a method where computers learn from data.",
    "Deep learning is a subset of machine learning that uses neural networks.",
    "Natural language processing helps computers understand human language.",
    "Computer vision enables machines to interpret images and videos.",
    "Reinforcement learning is a type of machine learning based on rewards."
]
# Load Embedding Model
print("Loading model...")
model = SentenceTransformer('all-MiniLM-L6-v2')
# Create Document Embeddings
print("Generating document embeddings...")
embeddings = model.encode(DOCUMENTS)
# Batch Retrieval Function
def batch_retrieve(queries, documents, embeddings, model, top_k=3):
    """
    Efficiently retrieve documents for multiple queries.
    """
    results_dict = {}

    # Create embeddings for all queries at once
    query_embeddings = model.encode(queries)

    # Normalize document embeddings
    doc_norms = np.linalg.norm(embeddings, axis=1)

    for query, query_embedding in zip(queries, query_embeddings):

        query_norm = np.linalg.norm(query_embedding)

        # Compute cosine similarity (vectorized)
        similarities = np.dot(embeddings, query_embedding) / (
            doc_norms * query_norm
        )

        # Get top_k indices
        top_indices = np.argsort(similarities)[::-1][:top_k]

        # Store results
        results = [(documents[i], similarities[i]) for i in top_indices]
        results_dict[query] = results

    return results_dict
# Testing  Batch Retrieval
queries = [
    "What is machine learning?",
    "Explain neural networks",
    "Tell me about reinforcement learning",
    "Tell me about quantum computing"  # Out-of-domain
]

batch_results = batch_retrieve(queries, DOCUMENTS, embeddings, model, top_k=2)
# Display Results
for query, results in batch_results.items():
    print("\nQuery:", query)

    for i, (doc, score) in enumerate(results, 1):
        print(f"Result {i} (Score: {score:.4f}):")
        print(doc)
        print()
````

````python
# DEFINE DOCUMENTS (Knowledge Base)
DOCUMENTS = [
    "Machine learning is a method where computers learn from data.",
    "Deep learning is a subset of machine learning that uses neural networks.",
    "Natural language processing helps computers understand human language.",
    "Computer vision enables machines to interpret images and videos.",
    "Reinforcement learning is a type of machine learning based on rewards."
]
# IMPORT REQUIRED LIBRARIES
import numpy as np
from sentence_transformers import SentenceTransformer
import warnings
import os
# Suppress unnecessary warnings and progress bars
warnings.filterwarnings("ignore")
os.environ["TRANSFORMERS_VERBOSITY"] = "error"
os.environ["HF_HUB_DISABLE_PROGRESS_BARS"] = "1"
# RETRIEVAL PIPELINE CLASS
class RetrievalPipeline:
    # Constructor → Loads embedding model
    def __init__(self, model_name='all-MiniLM-L6-v2'):
        print("Initializing RetrievalPipeline...")
        self.model = SentenceTransformer(model_name)  # Load embedding model
        self.documents = []                          # Store documents
        self.embeddings = None                       # Store document embeddings
        print("Model loaded.")

    # Add documents and generate embeddings
    def add_documents(self, documents):
        print(f"Adding {len(documents)} documents...")
        self.documents.extend(documents)             # Store documents
        self._generate_embeddings()                  # Create embeddings
        print("Documents added and embeddings generated.")

    # Private method → Generate embeddings for documents
    def _generate_embeddings(self):
        if self.documents:
            self.embeddings = self.model.encode(self.documents)
        else:
            self.embeddings = None

    # Private method → Compute cosine similarity
    def _cosine_similarity(self, vec1, vec2):
        return np.dot(vec1, vec2) / (
            np.linalg.norm(vec1) * np.linalg.norm(vec2)
        )

    # Retrieve top_k most similar documents
    def retrieve(self, query, top_k=3):

        # Check if documents exist
        if self.embeddings is None or not self.documents:
            raise ValueError("No documents added. Call add_documents() first.")

        # Convert query into embedding
        query_embedding = self.model.encode(query)

        similarities = []

        # Compute similarity with each document
        for i, doc_embedding in enumerate(self.embeddings):
            score = self._cosine_similarity(query_embedding, doc_embedding)
            similarities.append((self.documents[i], score))

        # Sort by highest similarity
        similarities.sort(key=lambda x: x[1], reverse=True)

        # Return top_k results
        return similarities[:top_k]

    # Return pipeline statistics
    def get_stats(self):
        return {
            "num_documents": len(self.documents),
            "embedding_dimension": self.embeddings.shape[1]
                if self.embeddings is not None else 0
        }
# TEST THE PIPELINE
# Initialize pipeline
pipeline = RetrievalPipeline()
# Add documents to pipeline
pipeline.add_documents(DOCUMENTS)
# Print statistics
print("Pipeline Statistics:")
print(pipeline.get_stats())
# Test query
query = "What are neural networks?"
results = pipeline.retrieve(query, top_k=2)
# Print results
print(f"\nQuery: {query}")
for doc, score in results:
    print(f"Score: {score:.4f} | {doc[:80]}...")
````

````python
# RELEVANCE METRICS
def calculate_precision_at_k(retrieved_docs, relevant_docs, k=3):
    retrieved_at_k = retrieved_docs[:k]
    relevant_count = sum(1 for doc in retrieved_at_k if doc in relevant_docs)
    return relevant_count / k if k > 0 else 0.0
def calculate_recall_at_k(retrieved_docs, relevant_docs, k=3):
    retrieved_at_k = retrieved_docs[:k]
    relevant_count = sum(1 for doc in retrieved_at_k if doc in relevant_docs)
    return relevant_count / len(relevant_docs) if len(relevant_docs) > 0 else 0.0
# Testing examples
retrieved_docs = [1, 0, 4]   # Example retrieved indices
relevant_docs = {0, 1}       # Ground truth relevant docs

precision = calculate_precision_at_k(retrieved_docs, relevant_docs, k=3)
recall = calculate_recall_at_k(retrieved_docs, relevant_docs, k=3)

print("Retrieved Docs:", retrieved_docs)
print("Relevant Docs:", relevant_docs)
print(f"Precision@3: {precision:.4f}")
print(f"Recall@3: {recall:.4f}")
````

### Challenge 1: Implement Re-ranking

After initial retrieval, re-rank results using a different strategy (e.g., keyword matching combined with semantic similarity).

````python
import numpy as np
import re
# Simple keyword overlap scorer
def keyword_overlap_score(query, document):
    # Convert to lowercase and tokenize
    query_words = set(re.findall(r"\w+", query.lower()))
    doc_words = set(re.findall(r"\w+", document.lower()))
    if len(query_words) == 0:
        return 0.0
    overlap = query_words.intersection(doc_words)
    return len(overlap) / len(query_words)
# Re-ranking function
def retrieve_with_reranking(query, pipeline, top_k=3,
                             semantic_weight=0.7, keyword_weight=0.3):

#Initial semantic retrieval (get more than needed)
    initial_results = pipeline.retrieve(query, top_k=top_k * 2)

    reranked_results = []

    for doc, semantic_score in initial_results:

# Compute keyword overlap
        keyword_score = keyword_overlap_score(query, doc)
#Combine scores
        final_score = (
            semantic_weight * semantic_score +
            keyword_weight * keyword_score
        )

        reranked_results.append((doc, final_score))
#Sort by combined score
    reranked_results.sort(key=lambda x: x[1], reverse=True)

    return reranked_results[:top_k]
# TEST RE-RANKING
query = "Explain neural networks in deep learning"
results = retrieve_with_reranking(query, pipeline, top_k=3)
print("Query:", query)
print("\nRe-ranked Results:")

for i, (doc, score) in enumerate(results, 1):
    print(f"\nResult {i}")
    print(f"Final Score: {score:.4f}")
    print(doc[:150] + "...")
````

### Challenge 2: Add Metadata Filtering

Extend your pipeline to support filtering documents by metadata (e.g., date, category, author).


````python
# METADATA-AWARE RAG RETRIEVAL SYSTEM (FULL WORKING CODE)
import numpy as np
from sentence_transformers import SentenceTransformer
#DOCUMENTS WITH METADATA
DOCUMENTS_WITH_METADATA = [
    {
        "text": "Machine learning enables systems to learn from data.",
        "category": "AI",
        "author": "Alice",
        "year": 2022
    },
    {
        "text": "Deep learning uses neural networks for complex tasks.",
        "category": "AI",
        "author": "Bob",
        "year": 2023
    },
    {
        "text": "Computer vision focuses on image recognition.",
        "category": "Vision",
        "author": "Alice",
        "year": 2021
    },
    {
        "text": "Natural language processing enables chatbots.",
        "category": "NLP",
        "author": "Charlie",
        "year": 2023
    }
]
# METADATA RETRIEVAL PIPELINE
class MetadataRetrievalPipeline:

    def __init__(self, model_name='all-MiniLM-L6-v2'):
        print("Initializing model...")
        self.model = SentenceTransformer(model_name)
        self.documents = []
        self.embeddings = None

    def add_documents(self, documents):
        self.documents.extend(documents)
        texts = [doc["text"] for doc in self.documents]
        print("Encoding documents...")
        self.embeddings = self.model.encode(texts)
        print("Documents encoded successfully!\n")
    def _cosine_similarity(self, vec1, vec2):
        return np.dot(vec1, vec2) / (
            np.linalg.norm(vec1) * np.linalg.norm(vec2)
        )
    def retrieve(self, query, top_k=3, filters=None):
        """
        Retrieve documents with optional metadata filtering
        """
        if self.embeddings is None:
            print("No documents added yet!")
            return []
        print(f"\nQuery: {query}")
        print(f"Filters: {filters}\n")
        query_embedding = self.model.encode(query)
        similarities = []
        for i, doc in enumerate(self.documents):
            # Apply Metadata Filtering
            if filters:
                match = True
                for key, value in filters.items():
                    if doc.get(key) != value:
                        match = False
                        break
                if not match:
                    continue
            # Compute Similarity
            score = self._cosine_similarity(
                query_embedding,
                self.embeddings[i]
            )

            similarities.append((doc, score))
        # Sort by score (highest first)
        similarities.sort(key=lambda x: x[1], reverse=True)

        return similarities[:top_k]
#  RUN THE SYSTEM
if __name__ == "__main__":

    # Initialize pipeline
    pipeline = MetadataRetrievalPipeline()

    # Add documents
    pipeline.add_documents(DOCUMENTS_WITH_METADATA)
 # Example 1: No Filter
    results = pipeline.retrieve("neural networks", top_k=3)
    print("Results (No Filter):")
    for doc, score in results:
        print(f"Score: {score:.4f}")
        print("Text:", doc["text"])
        print("Metadata:", {k:v for k,v in doc.items() if k != "text"})
        print("-" * 50)
  # Example 2: Filter by Category
    results = pipeline.retrieve(
        "neural networks",
        top_k=3,
        filters={"category": "AI"}
    )

    print("\nResults (Filter: category = AI):")
    for doc, score in results:
        print(f"Score: {score:.4f}")
        print("Text:", doc["text"])
        print("Metadata:", {k:v for k,v in doc.items() if k != "text"})
        print("-" * 50)
  # Example 3: Filter by Author
    results = pipeline.retrieve(
        "image recognition",
        top_k=3,
        filters={"author": "Alice"}
    )
    print("\nResults (Filter: author = Alice):")
    for doc, score in results:
        print(f"Score: {score:.4f}")
        print("Text:", doc["text"])
        print("Metadata:", {k:v for k,v in doc.items() if k != "text"})
        print("-" * 50)
````

### Challenge 3: Implement Hybrid Search

Combine dense retrieval (embeddings) with sparse retrieval (keyword-based like BM25) for improved results.

````python
# HYBRID SEARCH (DENSE + SPARSE)
import numpy as np
import re
from collections import Counter
from sentence_transformers import SentenceTransformer
# SAMPLE DOCUMENTS
DOCUMENTS = [
    "Machine learning enables systems to learn from data.",
    "Deep learning uses neural networks for complex tasks.",
    "Computer vision focuses on image recognition.",
    "Natural language processing enables chatbots.",
    "Transformers like BERT and GPT are used in NLP."
]
# ----------------------------------------------------------
# HYBRID RETRIEVAL PIPELINE
# ----------------------------------------------------------

class HybridRetrievalPipeline:

    def __init__(self, model_name='all-MiniLM-L6-v2'):
        print("Initializing Hybrid Retrieval System...")
        self.model = SentenceTransformer(model_name)
        self.documents = []
        self.embeddings = None
        self.tokenized_docs = []

    # ------------------------------------------------------
    # Add Documents
    # ------------------------------------------------------

    def add_documents(self, documents):
        self.documents = documents

        # Dense embeddings
        print("Encoding documents (Dense)...")
        self.embeddings = self.model.encode(documents)

        # Sparse tokenization
        print("Tokenizing documents (Sparse)...")
        self.tokenized_docs = [
            re.findall(r"\w+", doc.lower()) for doc in documents
        ]

        print("System ready!\n")
    # ------------------------------------------------------
    # Cosine Similarity (Dense)
    # ------------------------------------------------------

    def _cosine_similarity(self, vec1, vec2):
        return np.dot(vec1, vec2) / (
            np.linalg.norm(vec1) * np.linalg.norm(vec2)
        )
    # ------------------------------------------------------
    # Sparse Keyword Score (Simple TF matching)
    # ------------------------------------------------------

    def _keyword_score(self, query_tokens, doc_tokens):
        doc_counter = Counter(doc_tokens)
        score = 0

        for token in query_tokens:
            score += doc_counter[token]

        return score
    # ------------------------------------------------------
    # Hybrid Retrieve
    # ------------------------------------------------------

    def retrieve(self, query, top_k=3, alpha=0.7):
        """
        alpha → weight for dense score
        (1 - alpha) → weight for sparse score
        """

        query_embedding = self.model.encode(query)
        query_tokens = re.findall(r"\w+", query.lower())

        results = []

        for i, doc in enumerate(self.documents):

            # Dense Score
            dense_score = self._cosine_similarity(
                query_embedding,
                self.embeddings[i]
            )

            # Sparse Score
            sparse_score = self._keyword_score(
                query_tokens,
                self.tokenized_docs[i]
            )

            # Normalize sparse score
            sparse_score = sparse_score / (len(self.tokenized_docs[i]) + 1)

            # Hybrid Score
            hybrid_score = alpha * dense_score + (1 - alpha) * sparse_score

            results.append((doc, hybrid_score, dense_score, sparse_score))

        # Sort by hybrid score
        results.sort(key=lambda x: x[1], reverse=True)

        return results[:top_k]
# ----------------------------------------------------------
# RUN HYBRID SEARCH
# ----------------------------------------------------------
if __name__ == "__main__":

    pipeline = HybridRetrievalPipeline()
    pipeline.add_documents(DOCUMENTS)

    query = "Explain neural networks in deep learning"

    results = pipeline.retrieve(query, top_k=3, alpha=0.7)

    print("Query:", query)
    print("=" * 60)

    for doc, hybrid, dense, sparse in results:
        print(f"Hybrid Score: {hybrid:.4f}")
        print(f"Dense Score : {dense:.4f}")
        print(f"Sparse Score: {sparse:.4f}")
        print("Document:", doc)
        print("-" * 60)
````
