````python
### Task 2.2: Slicing operations (8 points)

Given `arr = np.arange(0, 20)`:

1. Extract elements from index 5 to 15 (inclusive)
2. Extract every 3rd element from the array
3. Reverse the entire array using slicing

Create a 4x4 2D array and:
- Extract the first 2 rows and first 3 columns
- Extract the last 2 rows and last 2 columns
- Extract every other row with all columns
````

````python
import numpy as np
arr = np.arange(0,20)
print("Original array:\n",arr)
#Extract elements from index 5 to 15
slice1 = arr[5:16]
print("\n Elements from index 5 to 15:",slice1)
# Extracting every 3rd element from the array 
slice2 = arr [::3]
print("\n Every 3rd element:",slice2)
#Reverse the entire array using slicing
slice3 = arr[::-1]
print("\n Reverse array :",slice3)
#Create a 4x4 2d array
arr2d = np.arange(1,17).reshape(4,4)
print("\n4x4 2D array:\n",arr2d)
#Extract the first 2 rows and 1ast 3 columns
slice4 = arr2d[:2,:3]
print("\n Firat 2 rows and last 3 columns :",slice4)
#Extract the last 2 rows and last 2 olumns
slice5 = arr2d[-2:,-2:]
print(" Last 2 rows and Last 2 columns:",slice5)
#Extract every other row with all columns
slice6 = arr2d[::2,:]
print("\n Every other row with all columns: ",slice6)
````

````python
### Task 3.2: Universal functions (7 points)

Given `arr = np.array([1, 4, 9, 16, 25, 36])`:

1. Calculate square root of all elements
2. Calculate exponential (e^x) of all elements
3. Calculate natural logarithm of all elements

Create an array of angles `[0, π/6, π/4, π/3, π/2]` and:
- Calculate sine values
- Calculate cosine values

For the array `[1.23, 2.56, 3.89, 4.12]`:
- Round to nearest integer
- Apply floor operation
- Apply ceiling operation
````

````python
import numpy as np
# Given array
arr = np.array([1, 4, 9, 16, 25, 36])
print("Original array:", arr)
# Square root of all elements
sqrt_arr = np.sqrt(arr)
print("\nSquare root:", sqrt_arr)
#Exponential (e^x) of all elements
exp_arr = np.exp(arr)
print("Exponential (e^x):", exp_arr)
#Natural logarithm of all elements
log_arr = np.log(arr)
print("Natural logarithm:", log_arr)
# Array of angles (in radians)
angles = np.array([0, np.pi/6, np.pi/4, np.pi/3, np.pi/2]) 
print("\nAngles (radians):\n", angles)
# Sine and Cosine values
sin_values = np.sin(angles)
cos_values = np.cos(angles)
print("Sine values:\n", sin_values)
print("Cosine values:\n", cos_values)
# Given decimal array
dec_arr = np.array([1.23, 2.56, 3.89, 4.12])
print("\nDecimal array:", dec_arr)
# Rounding operations
round_values = np.round(dec_arr)
floor_values = np.floor(dec_arr)
ceil_values = np.ceil(dec_arr)
print("Rounded values:", round_values)
print("Floor values:", floor_values)
print("Ceiling values:", ceil_values)
````

````python
### Task 4.2: Stacking and splitting (10 points)

Create three 1D arrays: `[1, 2, 3]`, `[4, 5, 6]`, `[7, 8, 9]`

1. Stack them vertically to create a 3x3 array
2. Stack them horizontally to create a 1D array

Create an array from 1 to 12 and:
- Split it into 4 equal parts

Create a 4x4 array with values 1-16 and:
- Split it horizontally into 2 parts
````

````python
import numpy as np
# Create three 1D arrays
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
c = np.array([7, 8, 9])
#Stack vertically to create a 3x3 array
vertical_stack = np.vstack((a, b, c))
print("Vertical stacking (3x3):\n", vertical_stack)
#Stack horizontally to create a 1D array
horizontal_stack = np.hstack((a, b, c))
print("\nHorizontal stacking (1D):\n", horizontal_stack)
# Create an array from 1 to 12
arr = np.arange(1, 13)
print("\nArray from 1 to 12:", arr)
# Split into 4 equal parts
split_arr = np.split(arr, 4)
print("Split into 4 equal parts:", split_arr)
# Create a 4x4 array with values 1 to 16
arr2d = np.arange(1, 17).reshape(4, 4)
print("\n4x4 array:\n", arr2d)
# Split horizontally into 2 parts
horizontal_split = np.hsplit(arr2d, 2)
print("Horizontal split into 2 parts:")
for part in horizontal_split:
    print(part)
````

````python
Exercise 6: Real-World Application - Student Grade Analysis (25 points)
You are given grade data for 5 students across 4 subjects (Math, Science, English, History):
python
grades = np.array([
    [85, 90, 78, 88],
    [92, 88, 95, 90],
    [76, 82, 80, 75],
    [88, 85, 90, 92],
    [95, 92, 88, 94]
]
**Tasks:**
1. **Calculate the average grade for each student** (5 points)
   - Output should show Student 1: X.XX, Student 2: Y.YY, etc.
2. **Calculate the average grade for each subject** (5 points)
   - Output: Math: X.XX, Science: Y.YY, etc.
3. **Find the highest and lowest grades in each subject** (5 points)
   - Output for each subject: Highest = X, Lowest = Y
4. **Identify students with an average grade above 85** (3 points)
   - Print student number and their average
5. **Calculate the overall class average** (2 points)
   - Average of all grades across all students
6. **Normalize all grades to a scale of 0-10** (5 points)
   - Use min-max normalization formula: (value - min) / (max - min) * 10
   - Print the normalized grades array
**Bonus:** Identify the top-performing student (+2 points)
````

````python
import numpy as np
grades = np.array([
    [85, 90, 78, 88],
    [92, 88, 95, 90],
    [76, 82, 80, 75],
    [88, 85, 90, 92],
    [95, 92, 88, 94]
])
subjects = ["Math", "Science", "English", "History"]
# ---------------- Task 1 ----------------
print("\n" + "=" * 60)
print("\n Task 1: Student Averages ")
print("\n" + "=" * 60)
student_avg = np.mean(grades, axis=1)
for i, avg in enumerate(student_avg, start=1):
    print(f"Student {i}: {avg:.2f}")
# ---------------- Task 2 ----------------
print("\n" + "=" * 60)
print("\n Task 2: Subject Averages")
print("\n" + "=" * 60)
subject_avg = np.mean(grades, axis=0)
for subject, avg in zip(subjects, subject_avg):
    print(f"{subject}: {avg:.2f}")
# ---------------- Task 3 ----------------
print("\n" + "=" * 60)
print("\n Task 3: Highest and Lowest Grades per Subject ")
print("\n" + "=" * 60)
subject_max = np.max(grades, axis=0)
subject_min = np.min(grades, axis=0)
for i, subject in enumerate(subjects):
    print(f"{subject}: Highest = {subject_max[i]}, Lowest = {subject_min[i]}")
# ---------------- Task 4 ----------------
print("\n" + "=" * 60)
print("\n Task 4: Students with Average > 85 ")
print("\n" + "=" * 60)
for i, avg in enumerate(student_avg, start=1):
    if avg > 85:
        print(f"Student {i}: {avg:.2f}")
# ---------------- Task 5 ----------------
print("\n" + "=" * 60)
print("\nTask 5: Overall Class Average")
print("\n" + "=" * 60)
overall_avg = np.mean(grades)
print(f"Overall Class Average: {overall_avg:.2f}")
# ---------------- Task 6 ----------------
print("\n" + "=" * 60)
print("\nTask 6: Normalized Grades (0–10 Scale)")
print("\n" + "=" * 60)
min_val = np.min(grades)
max_val = np.max(grades)
normalized_grades = (grades - min_val) / (max_val - min_val) * 10
print(np.round(normalized_grades, 2))
# ---------------- Bonus ----------------
print("\n--- Bonus: Top-Performing Student ---")
top_student = np.argmax(student_avg) + 1
print(f"Top-performing student: Student {top_student}")
````

````python
## Exercise 7: Challenge - Sales Data Analysis (30 points)

You are analyzing sales data for a company with 4 products over 6 months:

```python
sales_data = np.array([
    [120, 135, 150, 145, 160, 175],  # Product A
    [90, 95, 100, 110, 115, 120],    # Product B
    [150, 145, 155, 160, 170, 180],  # Product C
    [80, 85, 82, 88, 92, 95]         # Product D
])
```

**Tasks:**

1. **Calculate total sales for each product** (5 points)
   - Sum across all months for each product

2. **Calculate total sales for each month** (5 points)
   - Sum across all products for each month (Jan to Jun)

3. **Identify the month with the highest total sales** (4 points)
   - Print the month name and total sales

4. **Calculate the growth rate for each product** (6 points)
   - Growth rate = ((Last month - First month) / First month) × 100
   - Print as percentage for each product

5. **Find products with average monthly sales greater than 120k** (4 points)
   - Calculate average for each product
   - Print products meeting the criteria

6. **Calculate quarter-wise sales** (6 points)
   - Q1: January-March
   - Q2: April-June
   - Print sales for each product in each quarter

**Bonus:** Predict next month (July) sales for each product (+5 points)
- Use a simple method: average of last 2 months + average monthly growth
````

````python
import numpy as np
# Sales data
sales_data = np.array([
    [120, 135, 150, 145, 160, 175],  # Product A
    [90, 95, 100, 110, 115, 120],    # Product B
    [150, 145, 155, 160, 170, 180],  # Product C
    [80, 85, 82, 88, 92, 95]         # Product D
])
products = ["Product A", "Product B", "Product C", "Product D"]
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
#Total sales for each product
print("\n" + "=" * 60)
print("\n--- Total Sales for Each Product ---")
print("\n" + "=" * 60)
product_total = np.sum(sales_data, axis=1)

for i in range(len(products)):
    print(products[i], ":", product_total[i])
# ------------------------------------------------
# 2. Total sales for each month
# ------------------------------------------------
print("\n" + "=" * 60)
print("\n Total Sales for Each Month")
print("\n" + "=" * 60)
month_total = np.sum(sales_data, axis=0)
for i in range(len(months)):
    print(months[i], ":", month_total[i])
# ------------------------------------------------
# 3. Month with highest total sales
# ------------------------------------------------
print("\n" + "=" * 60)
max_month_index = np.argmax(month_total)
print("\n Month with Highest Sales ")
print(months[max_month_index], "with sales", month_total[max_month_index])
# ------------------------------------------------
# 4. Growth rate for each product
# ------------------------------------------------
print("\n" + "=" * 60)
print("\n--- Growth Rate for Each Product (%) ---")
print("\n" + "=" * 60)
for i in range(len(products)):
    first = sales_data[i][0]
    last = sales_data[i][-1]
    growth = ((last - first) / first) * 100
    print(products[i], ":", round(growth, 2), "%")
# ------------------------------------------------
# 5. Products with average monthly sales > 120
# ------------------------------------------------
print("\n" + "=" * 60)
print("\n--- Products with Average Sales > 120 ---")
print("\n" + "=" * 60)
product_avg = np.mean(sales_data, axis=1)

for i in range(len(products)):
    if product_avg[i] > 120:
        print(products[i], "Average:", round(product_avg[i], 2))
# ------------------------------------------------
# 6. Quarter-wise sales
# ------------------------------------------------
print("\n" + "=" * 60)
print("\n--- Quarter-wise Sales ---")
print("\n" + "=" * 60)
# Q1 = Jan, Feb, Mar (0 to 2)
# Q2 = Apr, May, Jun (3 to 5)

for i in range(len(products)):
    q1 = np.sum(sales_data[i, 0:3])
    q2 = np.sum(sales_data[i, 3:6])
    print(products[i], "Q1:", q1, "Q2:", q2)
# ------------------------------------------------
# Bonus task : Predict July sales
# ------------------------------------------------
print("\n" + "=" * 60)
print("\n--- Predicted July Sales ---")
print("\n" + "=" * 60)
for i in range(len(products)):
    last_two_avg = (sales_data[i][-1] + sales_data[i][-2]) / 2
    monthly_growth = (sales_data[i][-1] - sales_data[i][0]) / 5
    july_prediction = last_two_avg + monthly_growth
    print(products[i], "Predicted July Sales:", round(july_prediction, 2))
````
